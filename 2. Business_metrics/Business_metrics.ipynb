### 4. Профиль пользователя
Построим профиль пользователй, для каждого уникального пользователя запишем:

- acquisition_ts   - время первого захода
- acquisition_date - день первого захода 
- acquisition_month - месяц первого захода
- source_id - источник первого захода
- device - устройство первого захода
- acqusition_cost - стоимость привлечения пользователя
- first_purchase_dt - время первой покупки
- payer - является ли пользователь покупателем

```python
# Строим профили юзеров
users = (visits.sort_values(by = ['uid', 'start_ts'])
.groupby('uid')
.agg({'start_ts': 'first', 'source_id': 'first', 'device': 'first'})         
.rename(columns = {'start_ts' : 'acquisition_ts'})) 

users['acquisition_date'] = users['acquisition_ts'].dt.date
users['acquisition_month'] = users['acquisition_ts'].astype('datetime64[M]') 
users = (users.reset_index()
.set_index(['source_id','acquisition_date'])
.join(costs.set_index(['source_id', 'acquisition_date']), how = 'left'))

#добавляем стоимость приобретения пользователя
user_cost = (users.groupby(['source_id','acquisition_date'])
    .agg({'device': 'count', 'costs': 'max'})
    .rename(columns = {'device': 'users'})) 
user_cost['acqusition_cost'] = user_cost['costs'] / user_cost['users']

users = users.join(user_cost[['acqusition_cost']], how = 'left').reset_index()

users = users.set_index('uid')[['acquisition_ts','acquisition_date','acquisition_month',
                            'source_id', 'device', 'acqusition_cost']]

#платящий пользователь или нет
users = users.join(orders.groupby('uid').agg({'buy_ts':'min'}).rename(columns = {'buy_ts':'first_purchase_dt'}), how='left')
users['payer'] = ~users['first_purchase_dt'].isna() 
```

### 5. Средний чек
Рассчитаем средний чек пользователя по месячным когортам, по устройствам и по источникам привлечения.  
Время жизни пользователя возьмем 6 месяцев.

```python
# Напишем функцию для расчета среднего чека, на вход функция будет принимать когорту.

def user_rev(dimensions):

    revenue_per_user = orders.set_index('uid').join(users, how = 'left')[['acquisition_month','source_id', 'device',
                                                            'purchase_month','revenue']]

    revenue_per_user['age_month'] = ((revenue_per_user['purchase_month'] - revenue_per_user['acquisition_month']) / np.timedelta64(1,'M')).round().astype('int')
    revenue_per_user = revenue_per_user.reset_index()

    revenue_per_user = revenue_per_user.groupby(dimensions + ['age_month']).agg({'revenue':'sum','uid':'nunique'})
    
    revenue_per_user['revenue_per_user'] = revenue_per_user['revenue'] / revenue_per_user['uid']
    revenue_per_user = revenue_per_user.reset_index()
    revenue_per_user = revenue_per_user.pivot_table(index=dimensions, columns='age_month',values='revenue_per_user',aggfunc='mean')
    return revenue_per_user
   
```

Применим нашу функцию и построим графики по месячным когортам, по устройствам и по источникам привлечения.

```python
plt.figure(figsize=(10, 6))
plt.title('Средний чек покупателей')
sns.heatmap(user_rev(['acquisition_month']).iloc[0:7, 0:6], annot=True, fmt='.1f', linewidths=1, linecolor='gray')
plt.show()
```

```python
user_rev(['device']).iloc[0:7, 0:6].T.plot()
plt.title('Средний чек для разных устройств')
plt.xlabel('lifetime')
plt.ylabel('Чек')
plt.grid()
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
plt.show()
```

```python
user_rev(['source_id']).iloc[0:8, 0:6].T.plot()
plt.title('Средний чек для разныйх источников')
plt.ylabel('Чек')
plt.xlabel('lifetime')
plt.grid()
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
fig = plt.gcf()
fig.set_size_inches(8, 6)
plt.show()
```

Большинство когорт показывает рост среднего чека. Июньская когорта показывает лучше реззультаты, средний чек у нее выше в каждом месяце, чем у любой другой когорты.

### 7. Сколько раз покупают за период?

Рассчитаем сколько покупок делает пользователь по месячным когортам, по устройствам и по источникам привлечения.  
Время жизни пользователя возьмем 6 месяцев.

```python
# Напишем функцию для расчета количества покупок на одного пользователя, на вход функция будет принимать когорту.

def orders_per_user(dimensions):

    count_orders = orders.set_index('uid').join(users, how = 'left')[['acquisition_month','source_id', 'device',
                                                            'purchase_month', 'revenue']]

    count_orders['age_month'] = ((count_orders['purchase_month'] - count_orders['acquisition_month']) / np.timedelta64(1,'M')).round().astype('int')
    count_orders = count_orders.reset_index()

    count_orders = count_orders.groupby(dimensions + ['age_month']).agg({'revenue':'count','uid':'nunique'})
    
    count_orders['orders_per_user'] = count_orders['revenue'] / count_orders['uid']
    count_orders = count_orders.reset_index()
    count_orders = count_orders.pivot_table(index=dimensions, columns='age_month',values='orders_per_user',aggfunc='mean')
    return count_orders
```

Применим нашу функцию и построим графики по месячным когортам, по устройствам и по источникам привлечения.

```python
plt.figure(figsize=(10, 6))
plt.title('Количество покупок на пользователя')
sns.heatmap(orders_per_user(['acquisition_month']).iloc[0:7, 0:6], annot=True, fmt='.1f', linewidths=1, linecolor='gray')
plt.show()
```

### 8. Retention Rate
Рассчитаем процент удержания пользователей по месячным когортам, устройствам и источникам.

```python
def ret(dimensions):
    r_rate = orders.set_index('uid').join(users, how = 'left')[['acquisition_month','source_id', 'device',
                                                            'purchase_month','revenue']]

    r_rate['age_month'] = ((r_rate['purchase_month'] - r_rate['acquisition_month']) / np.timedelta64(1,'M')).round().astype('int')
    r_rate = r_rate.reset_index()

    r_rate = r_rate.groupby(dimensions + ['age_month']).agg({'uid':'nunique'}).reset_index()
    initial_users_count = r_rate[r_rate['age_month'] == 0][dimensions + ['uid']].rename(columns={'uid':'cohort_users'})
    r_rate = r_rate.merge(initial_users_count,on=dimensions)
    r_rate['retention'] = r_rate['uid'] / r_rate['cohort_users']

    retention_pivot = r_rate.pivot_table(index=dimensions, columns=['age_month'],values='retention',aggfunc='sum')
    
    return retention_pivot
```

Применим нашу функцию и построим графики по месячным когортам, по устройствам и по источникам привлечения.

```python
plt.figure(figsize=(10, 6))
plt.title('Удержание покупателей')
sns.heatmap(ret(['acquisition_month']).iloc[0:7, 0:6], annot=True, fmt='.3f', linewidths=1, linecolor='gray', vmax=0.2)
plt.show()
```

```python
plt.figure(figsize=(10, 4))
plt.title('Удержание покупателей для разных девайсов')
sns.heatmap(ret(['device']).iloc[0:7, 0:6], annot=True, fmt='.3f', linewidths=1, linecolor='gray', vmax=0.2)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.title('Удержание покупателей для разных источников')
sns.heatmap(ret(['source_id']).iloc[0:8, 0:6], annot=True, fmt='.3f', linewidths=1, linecolor='gray', vmax=0.2)
plt.show()
```

### 10. LTV / CAC / ROI

Рассчитаем:
- итоговую сумму денег, которую один клиент приносит в среднем со всех своих покупок.
- стоимость привлечения клиента
- возврат на инвестиции

Время жизни пользователя возьмем 6 месяцев.

```python
# Напишем функцию, которая будет рассчитывать размер когорты, стоимость привлечения всей когорты, стоимость привлечения клиента.
def cal_cohorts(dimensions):
    cohorts = (users.groupby(dimensions)
                   .agg ({'payer' : 'sum', 'acqusition_cost' : 'sum'})  #суммирует только payer ==True
                   .rename(columns = {'payer': 'cohort_size'})) 

    cohorts['cac'] = cohorts['acqusition_cost'] / cohorts['cohort_size'] 
    return cohorts

# Напишем функцию, которая будет рассчитывать ltv
def calc_ltv(dimensions):
    #считаем лтв
    ltv = orders.set_index('uid').join(users, how = 'left')[['acquisition_month','source_id', 'device',
                                                            'purchase_month','revenue']]
    ltv['age_month'] = ((ltv['purchase_month'] - ltv['acquisition_month']) / np.timedelta64(1,'M')).round().astype('int') 
    ltv = ltv.groupby(dimensions + ['age_month']).agg({'revenue': 'sum'})
    ltv = pd.pivot_table(ltv, index = dimensions, columns = 'age_month', values = 'revenue', aggfunc = 'sum')      
    ltv = ltv.div(cal_cohorts(dimensions)['cohort_size'], axis = 0) #делим все ячейки на соотвест когорту
    return ltv
```

Применим написанные функции и рассчитаем LTV средней когорты за 6 месяцев.   
Так же рассчитаем ROI, поделив LTV на CAC когорты.

```python
cohorts = cal_cohorts(['acquisition_month'])
ltv = calc_ltv(['acquisition_month']).round(2).iloc[0:7, 0:6]
ltv_mean = ltv.cumsum(axis=1).mean(axis=0)

roi = ltv.div(cohorts['cac'], axis = 0).cumsum(axis=1).round(2).iloc[0:7, 0:6]
roi_mean = roi.mean(axis=0)

mean_ltv_roi = pd.concat([ltv_mean, roi_mean], axis=1).rename(columns={0:'LTV', 1:'ROI'})
print('Средняя когорта LTV и ROI за каждый месяц их жизни')
mean_ltv_roi
```

```python
report = pd.concat([ltv.cumsum(axis=1).iloc[:, 5:6], roi.iloc[:, 5:6]], axis=1)
report.columns = ['LTV', 'ROI']
print('LTV и ROI на 6 месяц жизни когорты')
report
```
